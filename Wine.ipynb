```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.action_chains import ActionChains
```

```python
import time
path = "C:\Program Files (x86)\chromedriver.exe"
driver = webdriver.Chrome(path)
driver.get('https://www.majestic.co.uk/red-wine')

try:
    cookie = driver.find_element_by_xpath("//button[@id='onetrust-accept-btn-handler']")
    cookie.click()
except:
    pass
```

```python
import time
import pandas as pd
n = []
p = []

for i in range(20):
    print(i)
    names = driver.find_elements_by_xpath("//a[@class='product-name t-not-link']")
    for name in names:
        n.append(name.text)
    
    prices = driver.find_elements_by_xpath("//div[@class='product-price']")
    for price in prices:
        p.append(price.text)
        
    next_button = driver.find_element_by_xpath("//li[@class='next-page']")
    next_button.click()
    
    
    time.sleep(2)
driver.quit()   
df = pd.DataFrame({'Name':n,'Price':p})
df.to_csv('Red wine.csv')
print(df)
```
